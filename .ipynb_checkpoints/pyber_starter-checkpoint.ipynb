```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data = pd.read_csv("data/city_data.csv")
ride_data = pd.read_csv("data/ride_data.csv")

# Read the City and Ride Data

# Combine the data into a single dataset
#frames = [ride_data, city_data]
data = pd.merge(ride_data, city_data, on =('city'))
#data.sort_values('date')
# Display the data table for preview
data.head()
#cityType = data.type.nunique()
#cityType
```

```python
urban_fare_mean = data.loc[data["type"]=="Urban"].groupby("city").mean()["fare"]
suburban_fare_mean = data.loc[data["type"]=="Suburban"].groupby("city").mean()["fare"]
rural_fare_mean = data.loc[data["type"]=="Rural"].groupby("city").mean()["fare"]
```

```python
urban_driver_count = city_data.loc[data["type"]=="Urban"].groupby("city").count()["driver_count"]
suburban_driver_count = city_data.loc[data["type"]=="Suburban"].groupby("city").count()["driver_count"]
rural_driver_count = city_data.loc[data["type"]=="Rural"].groupby("city").count()["driver_count"]
print(rural_driver_count)
count = pd.DataFrame(rural_driver_count)
count.head()
```

## Total Fares by City Type

```python
######## Calculate Type Percents

# Provides the summation of all fares
totFare = data.fare.sum()  

# Provides the summation of fares based on city type
fareValueCounts = data.groupby(["type"])
fareValuetot = fareValueCounts.fare.sum()

# Converts to a dataframe 
dfFare = pd.DataFrame(fareValuetot)

# Locates the values for the fares based on the type of city
fareRural = dfFare.iloc[0]
fareSuburban = dfFare.iloc[1]
fareUrban = dfFare.iloc[2]

# Calculates the percentage of total fares by city type
percRuralFares = fareRural / totFare
percSuburbanFares = fareSuburban / totFare
percUrbanFares = fareUrban / totFare


pieFares = [percUrbanFares, percSuburbanFares, percRuralFares]


######### Build Pie Chart
plt.title("% of Total Fares by City Type")
fares = ["Urban", "Suburban", "Rural"]
colors = ["coral", "cyan", "yellow"]
explode = (0.1,0,0)
plt.pie(pieFares, explode=explode, labels=fares, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle = 270)
plt.show()
######### Save Figure
#plt.savefig("../Images/PyberFaresByCity2016.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

totDriver = city_data.driver_count.sum()
print(totDriver)
driverData= city_data.groupby(["type"])
dataType = driverData.driver_count.sum()
print(dataType)

dfDriver = pd.DataFrame(dataType)

driverRural = dfDriver.iloc[0]
driverSuburban = dfDriver.iloc[1]
driverUrban = dfDriver.iloc[2]

# Calculates the percentage of total fares by city type
percRuralDriver = driverRural / totDriver
percSuburbanDriver = driverSuburban / totDriver
percUrbanDriver = driverUrban / totDriver
print(percUrbanDriver)

# Build Pie Charts

pieDriver = [percUrbanDriver, percSuburbanDriver, percRuralDriver]

plt.title("% of Total Drivers by City Type")
rides = ["Urban", "Suburban", "Rural"]
colors = ["coral", "cyan", "yellow"]
explode = (0.1,0,0)
plt.pie(pieDriver, explode=explode, labels=rides, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle = 270)
plt.show()
######### Save Figure
#plt.savefig("../Images/PyberDriverByCity2016.png")
```
